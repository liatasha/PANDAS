```python
# Instanriating the encoder
one_hot_encoder = OneHotEncoder()
```

```python
# using the encoder to encode the MRI_Findings variable
from sklearn.preprocessing import OneHotEncoder


one_hot_encoder = OneHotEncoder(handle_unknown='ignore', sparse_output=False)

# Fit and transform the 'MRI_Findings' column to get the one-hot encoded features.

encoded_data = one_hot_encoder.fit_transform(brain_tumor[['MRI_Findings']])

# Create new column names for the one-hot encoded features
# These will be like 'MRI_Findings_Abnormal', 'MRI_Findings_Normal', 'MRI_Findings_Severe'
encoded_column_names = one_hot_encoder.get_feature_names_out(['MRI_Findings'])

# Create a DataFrame from the encoded data with appropriate column names and original index
encoded_df = pd.DataFrame(encoded_data, columns=encoded_column_names, index=brain_tumor.index)

# Add the new one-hot encoded columns to the 'brain_tumor' DataFrame
brain_tumor = pd.concat([brain_tumor, encoded_df], axis=1)

# Optionally, drop the original 'MRI_Findings' column now that it has been encoded into new columns
#brain_tumor.drop('MRI_Findings', axis=1, inplace=True)
```

```python

# Q2) What is the overall count and percentage of Brain Tumor Present (Yes/No)?

brain["Brain_Tumor_Present"] = brain["Brain_Tumor_Present"].map({"Yes":1, "No":0})
```

```python
# My solution
# 1a.
brain["Brain_Tumor_Present"].value_counts()
```

```python
# we added dropna=False so it will count null values if available
# 1a.
q2_counts = brain["Brain_Tumor_Present"].value_counts(dropna=False)

# normalize=True ensures the percentage falls between 0 or 1
# 1b.
q2_pct =  brain["Brain_Tumor_Present"].value_counts(normalize=True)*100

# \n is an escape character that allows the result to fall on the next line for legibility
print("\nQ2) Counts:\n", q2_counts)
print("\nQ2) Percent:\n", q2_pct.round(2))
```

```python
print("\nQ2) counts:\n", q2_counts)
print("\nQ2) Percent:\n", q2_pct.round(2))
```

```python
ax = q2_counts.plot(kind="bar")
ax.set_title("Q1: Brain Tumor Present (Counts)")
ax.set_xlabel("Brain_Tumor_Present")
ax.set_ylabel("Count")
plt.tight_layout()
plt.show()
```
